```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder # Dùng để mã hóa
from sklearn.naive_bayes import BernoulliNB # <--- QUAN TRỌNG: Dùng BernoulliNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import warnings

# Tắt các cảnh báo không quan trọng
warnings.filterwarnings('ignore')

# Cài đặt để Matplotlib hiển thị biểu đồ ngay trong notebook
%matplotlib inline
```

```python
# --- Bước 1: Tải và Khám phá Dữ liệu ---
print("--- Bước 1: Tải và Khám phá Dữ liệu ---")
try:
    data = pd.read_csv('./data/mushrooms.csv')
    print("Tải dữ liệu 'mushrooms.csv' thành công!")
    
    print("\n5 dòng dữ liệu đầu tiên:")
    display(data.head())
    
    print("\nThông tin chung về dữ liệu:")
    # .info() cho thấy tất cả đều là 'object' và không có null
    data.info()
    
except FileNotFoundError:
    print("Lỗi: Không tìm thấy tệp 'mushrooms.csv'.")
```

```python
# --- Bước 2: Tiền xử lý dữ liệu ---
# Vì Naive Bayes không thể xử lý dữ liệu chữ, chúng ta phải mã hóa TẤT CẢ các cột.

print("--- Bước 2: Bắt đầu Tiền xử lý dữ liệu ---")
data_processed = data.copy()

# 1. Mã hóa Biến mục tiêu (Target)
# Tách cột 'class' (y) ra trước
y = data_processed['class']
X = data_processed.drop('class', axis=1)

# Mã hóa y ('p' -> 1, 'e' -> 0)
le_y = LabelEncoder()
y_encoded = le_y.fit_transform(y)
print(f"Các lớp của y: {le_y.classes_} (đã mã hóa thành 0 và 1)")

# 2. Mã hóa Biến đặc trưng (Features)
# Sử dụng pd.get_dummies (One-Hot Encoding) cho TẤT CẢ các cột đặc trưng
X_encoded = pd.get_dummies(X, drop_first=True)

print(f"Đã mã hóa One-Hot Encoding cho các đặc trưng.")
print(f"Số lượng đặc trưng ban đầu: {len(X.columns)}")
print(f"Số lượng đặc trưng sau khi mã hóa: {len(X_encoded.columns)}")

print("\nDữ liệu đặc trưng (X) sau khi mã hóa (5 dòng đầu):")
display(X_encoded.head())
```

```python
# --- Bước 3: Chuẩn bị dữ liệu (Train/Test Split) ---
print("\n--- Bước 3: Chuẩn bị dữ liệu (Train/Test Split) ---")

# Chia dữ liệu (80% train, 20% test)
# (Giống Nhiệm vụ 1, bước 2 trong tài liệu)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)

print(f"Kích thước tập huấn luyện (X_train): {X_train.shape}")
print(f"Kích thước tập kiểm tra (X_test): {X_test.shape}")
```

```python
# --- Bước 4: Xây dựng và Huấn luyện mô hình Naive Bayes ---
print("\n--- Bước 4: Xây dựng và Huấn luyện mô hình ---")

# 1. Khởi tạo mô hình Bernoulli Naive Bayes
# (Vì dữ liệu One-Hot là 0/1, Bernoulli là phù hợp nhất)
classifier = BernoulliNB()

print("Đang huấn luyện mô hình BernoulliNB...")
# 2. Huấn luyện mô hình
classifier.fit(X_train, y_train)

print("Huấn luyện hoàn tất.")
```

```python
# --- Bước 5: Đánh giá mô hình Naive Bayes ---
print("\n--- Bước 5: Đánh giá mô hình trên tập Test ---")

# 1. Dự đoán trên tập X_test
# (Giống Nhiệm vụ 1, bước 5 trong tài liệu)
y_pred = classifier.predict(X_test)

# 2. Đánh giá Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# 3. In Báo cáo phân loại (Classification Report)
# (Giống Hình 2.7 trong tài liệu)
print("\nBáo cáo phân loại (Classification Report):")
# Lấy lại tên lớp gốc từ LabelEncoder (le_y)
class_names = [f"{cls} (decoded)" for cls in le_y.classes_]
print(classification_report(y_test, y_pred, target_names=class_names))

# 4. Vẽ Ma trận nhầm lẫn (Confusion Matrix)
# (Giống Hình 2.5)
print("\nMa trận nhầm lẫn (Confusion Matrix):")
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', 
            xticklabels=class_names, 
            yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Naive Bayes Confusion Matrix')
plt.show()
```
